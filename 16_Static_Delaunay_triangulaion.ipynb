# Триангуляция Делоне

Для начала определимся с терминологией.

>**Триангуляция** заданного набора точек $S$ - это такое разбиение выпуклой оболочки этих точек на треугольники, что множество вершин треугольников совпадает с $S$, любые два треугольника не пересекаются и любая точка из выпуклой оболочки принадлежит какому-нибудь треугольнику.

>**Критерий Делоне**: для любого треугольника нет точек из $S$, которые бы лежали внутри описанной около него окружности.

Собственно, **триангуляцией Делоне** назовем такую триангуляцию, для которой выполняется критерий Делоне

Кроме глобального китерия Делоне, определенного выше, существует локальный критерий Делоне, а также критерии Делоне для ребер.

>**Локальный критерий Делоне**: внутри окружности, описанной около треугольника, не лежат противолежащие ему вершины соседних треугольников.

>**Глобальный критерий Делоне для ребра**: на ребре можно построить такую окружность, что внутри неё не будет точек из $S$.

>**Локальный критерий Делоне для ребра**: на ребре можно построить такую окружность, что внутри неё не будет вершин, противолежащих ребру.

Необходимо доказать равносильность данных критериев.

### Лемма 1 (о равносильности критериев Делоне для треугольников)
> Локальный и глобальный критерии Делоне для треугольников равносильны

$\triangleright$

<div style="padding-left:40px">

Из глобального критерия очевидным образом следует локальный.<br>

Докажем в обратную сторону. Пусть для некоторого треугольника выполнен локальный, но не выполнен глобальный критерий. Т.е. существует точка $E$ такая, что лежит внутри описанной окружности и не имеет рёбер с вершинами треугольника. Зафиксируем эту точку и из всех таких треугольников рассмотрим тот, у которого $\angle BEC$ максимален, если $BC$ - ближайшая к точке $E$ сторона. Пусть с таким $\Delta ABC$ соседствует $\Delta BCD$. <br>

Точка $E$ не может лежать внутри $\Delta BCD$, так как в этом случае имелись бы рёбра $EB$ и $EC$, и нарушался бы локальный критерий. Значит, она снаружи. Не уменьшая общности, будем считать, что $CD$ - ближайшая к $E$ сторона $\Delta BCD$. Тогда выполнено, что $\angle BAC + \angle BEC > 180°, \angle BAC + \angle BDC < 180°$. Если т.$E$ лежит вне окружности, описанной около $\Delta BCD$, то $\angle BEC < \angle BDC$, что противоречит неравенствам выше. Значит, для $\Delta BCD$ также нарушается глобальный критерий. Из вышесказанного следует, что $\angle CED > \angle BEC$, где $CD$ - ближайшая к $E$ сторона. Противоречие.

</div>

$\triangleleft$

### Лемма 2 (о равносильности критериев Делоне для рёбер)
> Локальный и глобальный критерии Делоне для рёбер равносильны

$\triangleright$

<div style="padding-left:40px">

Из глобального критерия очевидным образом следует локальный.<br>

Докажем в обратную сторону. Пусть для некоторого ребра выполнен локальный, но не выполнен глобальный критерий. Т.е. существует такая точка, что лежит внутри любой описанной окружности и не образует треугольника с вершинами "плохого" ребра. Зафиксируем эту точку $D$ и из всех таких рёбер рассмотрим тот, у которого $\angle ADB$ максимален, где $AB$ - это ребро. Пусть это ребро со стороны точки $D$ лежит в $\Delta ABC$. <br>

Точка $D$ не может лежать внутри $\Delta ABC$, так как в этом случае имелись бы рёбра $AD$ и $BD$, и нарушался бы локальный критерий. Значит, она снаружи. Не уменьшая общности, будем считать, что $BC$ - ближайшая к $D$ сторона $\Delta ABC$. Относительно прямой $AD$ точки $B$ и $C$ лежат в разных полуплоскостях, а значит, $\angle BDC = \angle BDA + \angle ADC$. Отсюда $\angle BDC > \angle BDA$. Точка $D$ также не образует треугольника с ребром $BC$ и будет лежать в любой окружности, опирающейся на ребро $BC$ и не содержащей точку $A$, противолежащую ребру. Значит, при выборе ребра $AB$ мы должны были выбрать ребро $BC$. Противоречие.

</div>

$\triangleleft$

### Лемма 3 (о равносильности глобальных критериев Делоне)
> Глобальные критерии Делоне для треугольников и для рёбер равносильны

$\triangleright$

<div style="padding-left:40px">

Из критерия для треугольников очевидным образом следует критерий для рёбер. Достаточно в качестве необходимой окружности взять окружность одного из треугольников, содержащих противолежащую точку.<br>

Докажем в обратную сторону. Пусть для всех рёбер выполнен глобальный критерий Делоне, но есть $\Delta ABC$, чья описанная окружность содержит какую-либо точку $D$ из набора  $S$. Пусть $BC$ - ближайшее к точке $D$ ребро. Для него выполнено $\angle BAC + \angle BDC > 180°$. Из этого следует, что невозможно построить окружность, не содержащую одновременно точек $A$ и $D$. Значит, есть ребро, для которого не выполнен глобальный критерий Делоне. Противоречие.

</div>

$\triangleleft$

Из предыдущих трех лемм следует, что **все критерии Делоне равносильны.**
